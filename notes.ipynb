### What information each node in MCTS contains

1. Game State
    * what the board looks like
    * player to move
    * outcome of the game (if it has ended)
2. Connections to all possible moves
    * each possible move is its own node
3. Visit Couts ***N(s, a)***
    * number of times a node (or action) was taken
    * starts at 0
    * increments by 1 each time a node is selected
4. Value Head ***V(s, a)***
    * game state estimate by the neural network
    * metric for exploitation
5. Policy Head ***P(s, a)***
    * softmax vector of all possible moves and how promising each one is
    * metric for exploration
6. Total Value ***W(s, a)***
    * sum of all value heads that pass through this node
7. Mean Value ***Q(s, a) = W(s, a) / N(s, a)***
    * average value of this action
8. Meta Information:
    * is_root (used to know where backpropagation stops)
    * list of which possible moves have been added (important to know if a node has been fully expanded)
    * parent node (so the node knows who to backpropagate values to)
